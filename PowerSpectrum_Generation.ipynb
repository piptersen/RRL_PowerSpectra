### Power Spectrum for Single Lines ($\Delta n_{RRL} = 1$)
Generates a series of power spectrum, depending on the given parameters and lines for:
- 21cm auto-Power Spectrum
- RRL auto-Power Specturm (for n1)
- 21cm x RRL (n1) cross-Power Spectrum
- RRL (n1) x RRL (n2) cross-Power Spectrum

```python
import numpy as np
import matplotlib.pyplot as plt
import PS_calculation as psc
import importlib
importlib.reload(psc)
n1 = 160
n2 = 165

k_sin, P_21_sin, P_RRL_sin, P_21_RRL_sin, P_RRL_RRL_sin = psc.singlePS_calc(k_min = 5e-2, k_max = 10, N = 10_000, z_21 = 2, bandwidth = 0.05, 
                                                       n=n1, n_2=n2, b21 = 2, bRRL = 5, bRRL_2 = 5, RRL_separation = 50)
```

### Power Spectrum for $\Delta n_{RRL} > 1$
Generates a series of power spectrum, depending on the given parameters for:
- 21cm auto-Power Spectrum
- RRL auto-Power Specturm (for each $n_{RRL}$)
- 21cm x RRL cross-Power Spectrum
- RRL x RRL cross-Power Spectrum

The 21cm x RRL cross-Power spectrum is a combination of the cross power spectrum of 21cm with each line between $n_{min}$ and $n_{max}$.

The RRL x RRL cross-Power spectrum is a combination of the cross power spectrum of each RRL between $n_{min}$ and $n_{max}$ with each other RRL

```python
import numpy as np
import matplotlib.pyplot as plt
import PS_calculation as psc
import importlib
importlib.reload(psc)

n_min = 100
n_max = 200
k, P_21, P_RRL, P_21_RRL, P_RRL_RRL = psc.PS_calc_n_space(k_min = 5e-2, k_max = 10, N = 10_000, z_21 = 2, bandwidth = 0.2, 
                                                      n=n_min, n_2=n_max, b21 = 1, bRRL = 3, bRRL_2 = 3, 
                                                      RRL_separation = 50)
```

```python
plt.loglog(k, P_21, label = '21cm auto')#, c='blue')
plt.loglog(k, P_RRL, label = 'RRL auto')#, c = 'maroon')

plt.loglog(k, P_21_RRL, label = '21 x RRL', alpha = 0.9)#, c = 'tan')
plt.loglog(k, P_RRL_RRL, label = 'RRL x RRL', c = 'maroon', alpha = 0.6)#, c='purple')

#plt.xlim(5e-2, 1)
plt.xlabel(r'$\text{k}~\left[\text{h}~\text{Mpc}^{-1}\right]$')
plt.ylabel(r'$\Delta^2\left(\text{k}\right)$')
#plt.xlim(1e-2, 1)
plt.title(fr'Power spectra ($\tau = 0.1$, ${n_min} < n < {n_max}$)')
plt.legend()
```
